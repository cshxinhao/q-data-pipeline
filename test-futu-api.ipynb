# All Listed Stocks

```python
import pandas as pd

# Direct URL (no login needed)
url = "https://www.hkex.com.hk/eng/services/trading/securities/securitieslists/ListOfSecurities.xlsx"

# The file has title rows on top → skip them
df = pd.read_excel(url, skiprows=2, engine="openpyxl")

# Filter to only equity stocks (Main Board + GEM + others)
equity_df = df[df["Category"] == "Equity"].copy()

# Pad stock codes to 5 digits (official HKEX format)
equity_df["Stock Code"] = equity_df["Stock Code"].astype(str).str.zfill(5)

# Add yfinance-compatible ticker (e.g. 00001.HK)
equity_df["Ticker"] = equity_df["Stock Code"] + ".HK"

# Preview
print(
    equity_df[["Stock Code", "Name of Securities", "Sub-Category", "Ticker"]].head(10)
)

# Save for later use
equity_df.to_csv("hkex_equity_stocks.csv", index=False)
print(f"Saved {len(equity_df)} equity stocks to hkex_equity_stocks.csv")
```
